# Data preprocessing

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.nunique()
```

```python
df.describe()
```

```python
df['AMOUNT']=df['AMOUNT'].str.replace('$','',regex=True)
```

```python
def applyfun(d):
    if 'M' in d:
        return float(d.replace('M',''))
    else :
         return float(d.replace('B',''))*1000


df['AMOUNT'] = df['AMOUNT'].apply(applyfun)
```

```python
df.head(5)
```

```python
df.info()
```

```python
df['LENDER'].unique()
```

```python
l = []
for i in df['LENDER'].unique():
    l+=[j for j in i.split(', ')  if j not in l]
    
l
```

```python
df[l] = 0
```

```python
for i,j in enumerate(df['LENDER']):
    for g in j.split(', '):
        df[g][i] = 1
        
df.head(5)
```

```python
df['BORROWER'].unique()
```

```python
t = []

for i in df['BORROWER']:
    if 'Gov' in i :
        t.append('Government')
        
    elif "bank" in i.lower():
        t.append('Bank')
        
    else:
        t.append('Company')
```

```python
len(t)
```

```python
df['Borrower_type'] = t
```

```python
df.drop(['BORROWER','LENDER'],axis=1,inplace=True)
df.head()
```

```python
df['SECTOR'].unique()
```

```python
df['SECTOR'].value_counts()
```

```python
df['SECTOR'] = df['SECTOR'].replace('fTransport','Transport')
df['SECTOR'] = df['SECTOR'].replace('Extraction, pipelines','Extraction and pipelines')
df['SECTOR'] = df['SECTOR'].replace('Mulit-sector/discretionary','MultiSector and discretionary ')
```

```python
df['SECTOR'].value_counts()
```

```python
df['SENSITIVE_TERRITORY_OVERLAP'].value_counts()
```

```python
l = []
for i in df['SENSITIVE_TERRITORY_OVERLAP'].unique():
    l+=[j for j in i.split('\n ')  if j not in l]
    
l
```

```python
df[['STO_Indigenous','STO_Critical','STO_Protected']] = 0
```

```python
for i,j in enumerate(df['SENSITIVE_TERRITORY_OVERLAP']):
    for g in j.split('\n '):
        if  "Within Indigenous Peoples' Lands" in g:
            df['STO_Indigenous'][i] = 1 
            
        if  'Within Critical Habitats' in g:
            df['STO_Critical'][i] = 1 
            
        if  'Within National Protected Areas'  in g:
            df['STO_Protected'][i] = 1  
            
df.head(2)
```

## Realtion between columns

```python
amount = df['AMOUNT']
year = df['YEAR']
```

```python
cov_matrix = np.cov(amount, year)
cov_XY = cov_matrix[0,1]
print(f"Covariance between amount and year : {cov_XY}")
```

```python
corr_coefficient = np.corrcoef(amount, year)[0, 1]
print(f"Pearson corelation between amount and year : {corr_coefficient}")
```
